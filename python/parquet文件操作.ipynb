## 通过 DuckDB 操作 Parquet 文件

```python
import duckdb

## 创建或连接到 DuckDB
con = duckdb.connect(database=":memory:")

## 读取 Parquet 文件并加载到表中
con.execute(f"CREATE TABLE parquet_table AS SELECT * FROM read_parquet('{parquet_path}')")

## 执行查询 80 - 108 的数据，并将结果生成到一个新的 Parquet 文件中
con.execute(f"COPY (SELECT * FROM parquet_table where freq_start <= 80000000 and freq_end >= 108000000) TO  '/Users/h_vk/Downloads/test.parquet' (FORMAT PARQUET)")
```
