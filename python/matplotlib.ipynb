### Matplotlib 绘图标记

plot() 方法的 marker 参数来定义，绘图标记

- "."： 点
- ","：像素点
- "o"：实心圆
- "v"：下三角
- "^"：上三角
- "<"：左三角
- ">"：右三角

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([1,3,4,5,8,9,6,1,3,4,5,2,4])

plt.plot(ypoints, marker = 'o')
plt.show()
```

### fmt 参数
fmt 参数定义了基本格式，如标记、线条样式和颜色，格式为 [marker][line][color]，例如 o:r，表示 o 表示实心圆，: 表示虚线，r 表示红色。

line 类型：
- '-': 实线
- ':': 虚线
- '--': 破折线
- '-.': 点划线

color 类型：
- 'r': 红色
- 'g': 绿色
- 'b': 蓝色
- 'c': 青色
- 'm': 品红
- 'y': 黄色
- 'k': 黑色
- 'w': 白色



```python
import matplotlib.pyplot as plt
import numpy as np

data = np.array([ 2, 4, 6, 8 ])

plt.plot(data, 'o:r')
plt.show()
```

### 标记大小与颜色
自定义标记的大小与颜色，使用的参数分别是：
- markersize，简写为 ms：定义标记的大小。
- markerfacecolor，简写为 mfc：定义标记内部的颜色。
- markeredgecolor，简写为 mec：定义标记边框的颜色。

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.array([ 2, 4, 6, 8 ])

plt.plot(data, 'o:r', ms=20, mec='b')
plt.show()
```

### Matplotlib 轴标签和标题
通过 xlabel()、ylabel() 方法设置 x轴、y轴的标签。
通过 title() 方法设置图标标题

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = np.array([1, 4, 9, 16])

plt.plot(x, y, 'o:r', ms=20, mec='b')
plt.title('TEST Title')
plt.xlabel('x - label')
plt.ylabel('y - label')
plt.show()
```

### 图表中文支持

Matplotlib 默认情况不支持中文，可以通过指定字体

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = np.array([1, 4, 9, 16])

plt.plot(x, y, 'o:r', ms=20, mec='b')

# 自定义字体
# zhfont1 = matplotlib.font_manager.FontProperties(fname="SourceHanSansSC-Bold.otf", size=18)
# plt.title('TEST Title', fontproperties=zhfont1)

# 获取系统字体列表
# matplotlib.font_manager.fontManager.ttflist

# 指定系统字体
plt.rcParams['font.family'] = ['STFangsong']

plt.title('图表标题')
plt.xlabel('x 轴标签')
plt.ylabel('y 轴标签')
plt.show()
```

### 标题与标签的定位

title() 方法提供了 loc 参数来设置标题显示的位置，可以设置为: 'left', 'right', 和 'center'， 默认值为 'center'。

xlabel() 方法提供了 loc 参数来设置 x 轴显示的位置，可以设置为: 'left', 'right', 和 'center'， 默认值为 'center'。

ylabel() 方法提供了 loc 参数来设置 y 轴显示的位置，可以设置为: 'bottom', 'top', 和 'center'， 默认值为 'center'。

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = np.array([1, 4, 9, 16])

plt.plot(x, y, 'o:r', ms=20, mec='b')
plt.title('TEST Title', loc='left')
plt.xlabel('x - label', loc='left')
plt.ylabel('y - label', loc='top')
plt.show()
```
